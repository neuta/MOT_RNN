```python
def num_obj(Obj):
    str_obj = np.array(['DontCare', 'Van', 'Pedestrian', 'Cyclist', 'Car'])
    if str_obj[0] == Obj: 
        n = 0  
    elif str_obj[1] == Obj:
        n = 1
    elif str_obj[2] == Obj:
        n = 2
    elif str_obj[3] == Obj:
        n = 3
    elif str_obj[4] == Obj:
        n = 4    
    else:
        print('error')        
    return n
```

```python
def obj_conc(N_O, features, C_DontCare, C_Van, C_Pedestrian, C_Cyclist, C_Car):
    
    if N_O == 0:    
        C_DontCare = np.concatenate((C_DontCare, features), axis = 0)
    elif N_O == 1:
        C_Van = np.concatenate((C_Van, features), axis = 0)
    elif N_O == 2:
        C_Pedestrian = np.concatenate((C_Pedestrian, features), axis = 0)
    elif N_O == 3:
        C_Cyclist = np.concatenate((C_Cyclist, features), axis = 0)
    elif N_O == 4:
        C_Car = np.concatenate((C_Car, features), axis = 0)
    else:
        print('error')  
    
    return C_DontCare, C_Van, C_Pedestrian, C_Cyclist, C_Car
```

```python
def obj_reshape(C_DontCare, C_Van, C_Pedestrian, C_Cyclist, C_Car):
    c = 10

    r_D = int(C_DontCare.shape[0]/c)
    r_V = int(C_Van.shape[0]/c)
    r_P = int(C_Pedestrian.shape[0]/c)
    r_C = int(C_Cyclist.shape[0]/c)
    r_K = int(C_Car.shape[0]/c)

    print('DontCare'.center(60,'-'))
    C_DontCare = np.reshape(C_DontCare, (r_D, c))
    print(C_DontCare)
    print(C_DontCare.shape)

    print('Van'.center(60,'-'))
    C_Van = np.reshape(C_Van, (r_V, c))
    print(C_Van)
    print(C_Van.shape)

    print('Pedestrian'.center(60,'-'))
    C_Pedestrian = np.reshape(C_Pedestrian, (r_P, c))
    print(C_Pedestrian)
    print(C_Pedestrian.shape)


    print('Cyclist'.center(60,'-'))
    C_Cyclist = np.reshape(C_Cyclist, (r_C, c))  
    print(C_Cyclist)
    print(C_Cyclist.shape)
    
    print('Car'.center(60,'-'))
    C_Car = np.reshape(C_Car, (r_K, c))  
    print(C_Car)
    print(C_Car.shape)
    
    return C_DontCare, C_Van, C_Pedestrian, C_Cyclist, C_Car
```

## Others

```python
def general_reshape(data, N):    
    colums = N # Features
    rows = int(data.shape[0]/colums)
    new = np.reshape(data, (rows, colums))    
    return new
```

```python
def reshape_batch(track, N, F):
    batch = math.floor(track.shape[0]/N)
    tn = np.resize(track,(N*batch,F))
    new_data = np.reshape(tn,(batch,N,F))
    return new_data
```

```python
def general_reshape_labels(data, N):    
    colums = N # Features
    rows = int(data.shape[0]/colums)
    new = np.reshape(data, (colums, rows))    
    return new
```
